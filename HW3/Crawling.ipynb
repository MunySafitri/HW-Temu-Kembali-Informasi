```python
from tkinter import N
from turtle import title
import requests
from bs4 import BeautifulSoup

kolom1 = []
kolom2 = []
kolom3 = []
nomor =[] 
k=1
for i in range(10):
  url = 'https://www.hukumonline.com/klinik/pidana/page/'+str(i)+'/'
  #Membuat Requests
  r = requests.get(url)
  #Hasil Response
  request = r.content

  soup = BeautifulSoup(request, 'html.parser')
  #Extract Element
  title = soup.findAll('h2', attrs={'class':'header-6 font-weight-bold text-dark'})
  date = soup.findAll('span',attrs={ 'class':'small text-muted mr-1'})
  author = soup.findAll('a',attrs={'class':"small mx-2 d-none d-md-inline-block"})
  count = 0
  for x in range(0, len(title)):
    count += 1
    # print("{0}. {1}\n  {2}".format(count,title[x].text.strip(),date[x].text.strip()))
    kolom1.append(title[x].text.strip())
    kolom2.append(date[x].text.strip())
    kolom3.append(author[x].text.strip())
    nomor.append(str(k))
    k=k+1
```

```python
import numpy as np
data = {'nomor':nomor,
        'nama': kolom1,
        'update': kolom2,
        'author':kolom3
       }
# arr = np.array([kolom1,kolom2])
```

```python
import pandas as pd
df_hasil = pd.DataFrame(data)
```
